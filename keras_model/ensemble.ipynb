```python
import numpy as np
import pandas as pd


tfidf_lr = pd.read_csv('prob_lr_baseline.csv') # PL score 0.9829
testcnn = pd.read_csv('prob_textcnn_test.csv') # 0.9811

```

```python
tfidf_lr.head()
```

```python
testcnn.head()
```

```python
ensemble = tfidf_lr+testcnn
```

```python
ensemble.head()
```

```python
ensemble = np.array(ensemble.drop(['id'],axis=1))
```

```python
ensemble[:5]
```

```python
all_predictions=np.argmax(ensemble,axis=1)
```

```python
all_predictions = all_predictions+1
```

```python
all_predictions[:5]
```

```python
test_id = pd.read_csv("data/test_set.csv")
```

```python
test_pred=pd.DataFrame(all_predictions)
test_pred.columns=["class"]
test_pred["class"]=(test_pred["class"]).astype(int)
print(test_pred.shape)
print(test_id.shape)
test_pred["id"]=list(test_id["id"])
test_pred[["id","class"]].to_csv('ensemble.csv',index=None)
```
